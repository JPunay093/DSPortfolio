### Importing of the packages

```python
import yfinance as yf
import pandas as pd
```

### Using download() method to download the stock data of AMD

```python
amd = yf.download("AMD", start="1990-01-01", end = "2023-12-31", group_by="tickers")
```

Checking the data by using the print() method

```python
print(amd)
```

```python
amd.reset_index(inplace=True)
```

### Plotting of the stock prices from the start of the stock opening until the latest date of when the program was run.

```python
amd.plot(x="Date", y="Open")
```
